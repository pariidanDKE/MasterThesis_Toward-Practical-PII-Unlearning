#### Visualize results for my experiment regarding P, for Llama3.1-8B

```python
import pandas as pd

main_exp_path = '/projects/0/hpmlprjs/LLM/danp/UGBench/experiment/PII/UnlearnParamsExperiment/forget10/SubtractionCoefficientExperiment/llama3.1-8b'
all_results = None
Ps = [0.4, 0.8, 1.2, 1.6]

for run in range(1,6):
    for P in Ps:
        template_path = f'llama3.1-8b_E8_B16_C0.1_P_{P}_permu_intextFalse_run{run}_dosampleFalse'
        results_path = 'eval_results/eval.csv'
        full_path = main_exp_path + '/' + template_path + '/' + results_path
        df = pd.read_csv(full_path)
        df['P'] = P
        df['run'] = run
        if run == 1 and P == Ps[0]:
            all_results = df
        else:
            all_results = pd.concat([all_results, df], ignore_index=True)      

grouped_results = all_results.groupby(['P']).mean(numeric_only=True).reset_index()
```
